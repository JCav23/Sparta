```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
#check for missing values
titanic.isnull().sum()
```

```python
#only 202 values known out of 890 Passenger Deck Values so drop column
titanic.drop(labels='deck', axis=1, inplace=True)
titanic
```

```python
# Displaying passenger info by gender and check for null values within each group
male = titanic[titanic['sex'] == 'male']
female = titanic[titanic['sex'] == 'female']
female.isnull().sum()
```

```python
male.isnull().sum()
```

```python
#male_clean = male.interpolate(method='pad', axis=1)
#FutureWarning: DataFrame.interpolate with method=pad is deprecated and will raise in a future version. Use obj.ffill() or obj.bfill() instead.

# calculate average ages for each gender
avg_male_age = round(np.mean(male['age']))
avg_female_age = round(np.mean(female['age']))
# replace missing values with average female age
female['age'].fillna(avg_female_age, inplace=True)
male['age'].fillna(avg_male_age, inplace=True)
```

```python
# Concatenate the male & female results back together
titanic = pd.concat([male,female], axis=0)
titanic
```

```python
#Check for duplicate rows
duplicate_rows = titanic[titanic.duplicated(keep=False)]
duplicate_rows
```

```python
## Drop duplicated rows keeping last instance of duplicate
```

```python
titanic = titanic.drop_duplicates(keep='last', ignore_index=True)
titanic
```
